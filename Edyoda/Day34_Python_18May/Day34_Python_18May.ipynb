```python
#extract Image links

import requests
from bs4 import BeautifulSoup
	
def getdata(url):
	r = requests.get(url)
	return r.text
	
htmldata = getdata("https://en.wikipedia.org/wiki/Virat_Kohli")
soup = BeautifulSoup(htmldata, 'html.parser')
urls=[]
for item in soup.find_all('img'):
	
    s='https://'+item['src']
    urls.append(s) 
    print('https:'+item['src'])
print(urls,len(urls))    
```
